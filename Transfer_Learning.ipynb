### Descargar sets de datos

```python
DATA_PATH = '/CIFAR10'
NUM_TRAIN = 45000
MINIBATCH_SIZE = 64
transform_imagenet = T.Compose([
                T.Resize(224),
                T.ToTensor(),
                T.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))
            ])

transform_cifar = T.Compose([
                T.ToTensor(),
                T.Normalize([0.491, 0.482, 0.447], [0.247, 0.243, 0.261])
            ])

# Training set loader
cifar10_train = datasets.CIFAR10(DATA_PATH, train=True, download=True,
                             transform=transform_imagenet)
train_loader = DataLoader(cifar10_train, batch_size=MINIBATCH_SIZE,
                          sampler=sampler.SubsetRandomSampler(range(NUM_TRAIN)))

# Validation set loader
cifar10_val = datasets.CIFAR10(DATA_PATH, train=True, download=True,
                           transform=transform_imagenet)
val_loader = DataLoader(cifar10_val, batch_size=MINIBATCH_SIZE,
                        sampler=sampler.SubsetRandomSampler(range(NUM_TRAIN, len(cifar10_val))))

# Testing set loader
cifar10_test = datasets.CIFAR10(DATA_PATH, train=False, download=True,
                            transform=transform_imagenet)
test_loader = DataLoader(cifar10_test, batch_size=MINIBATCH_SIZE)
```

```python
for i, (x, y) in enumerate(val_loader):
    print(i, x.shape, y.shape)
```

### Usar GPU

```python
if torch.cuda.is_available():
    device = torch.device('cuda')
else:
    device = torch.device('cpu')

print(device)
```

### Mostrar imágenes

```python
classes = ['Plane', 'Car', 'Bird', 'Cat', 'Deer','Dog', 'Frog', 'Horse', 'Ship', 'Truck']

def plot_figure(image):
    plt.imshow(image.permute(1,2,0))
    plt.axis('off')
    plt.show()

rnd_sample_idx = np.random.randint(len(test_loader))
print(f'La imagen muestreada representa un: {classes[test_loader.dataset[rnd_sample_idx][1]]}')
image = test_loader.dataset[rnd_sample_idx][0]
image = (image - image.min()) / (image.max() -image.min() )
plot_figure(image)
```

### Calcular Accuracy

```python
def accuracy(model, loader):
    num_correct = 0
    num_total = 0
    model.eval()
    model = model.to(device=device)
    with torch.no_grad():
        for (xi, yi) in loader:
            xi = xi.to(device=device, dtype = torch.float32)
            yi = yi.to(device=device, dtype = torch.long)
            scores = model(xi) # mb_size, 10
            _, pred = scores.max(dim=1) #pred shape (mb_size )
            num_correct += (pred == yi).sum() # pred shape (mb_size), yi shape (mb_size, 1)
            num_total += pred.size(0)
        return float(num_correct)/num_total

```

### Loop de entrenamiento

```python
def train(model, optimiser, epochs=100):
#     def train(model, optimiser, scheduler = None, epochs=100):
    model = model.to(device=device)
    for epoch in range(epochs):
        for i, (xi, yi) in enumerate(train_loader):
            model.train()
            xi = xi.to(device=device, dtype=torch.float32)
            yi = yi.to(device=device, dtype=torch.long)
            scores = model(xi)

            cost = F.cross_entropy(input= scores, target=yi)

            optimiser.zero_grad()
            cost.backward()
            optimiser.step()

        acc = accuracy(model, val_loader)
#         if epoch%5 == 0:
        print(f'Epoch: {epoch}, costo: {cost.item()}, accuracy: {acc},')
#         scheduler.step()
```

#Aplicacion de MobileNet V3

```python
# Model definition
efficientnet = models.efficientnet_v2_m(weights=models.EfficientNet_V2_M_Weights.DEFAULT)
```

```python
# Visualize the parameter with required grad
for i, w in enumerate(efficientnet.parameters()):
    print(i, w.shape, w.requires_grad)
```

```python
p_efficientnet = nn.Sequential(*list(efficientnet.children()))
p_efficientnet
```

Como podemos observar, la ultima capa de la red es Fully Connected con 1000 neuronas de salida que representan las clases, lo que haremos sera sustituir esta capa por una capa FC con el numero de clases que necesitamos, en este caso 10

```python
# Get all layers but last one
p_efficientnet = nn.Sequential(*list(efficientnet.children()))[:-1]
p_efficientnet
```

```python
# Freeze network parameters
for i, parameter in enumerate(p_efficientnet.parameters()):
    parameter.requires_grad = False
```

```python
# Check out
for i, parameter in enumerate(p_efficientnet.parameters()):
    print(parameter.requires_grad)
```

```python
# Define hyperparamters and epochs
lr = 5e-4
epochs = 5

# Build network architecture including new FC layer with 10 output neurons
model2 = nn.Sequential(p_efficientnet,
                       nn.Flatten(),
                       nn.Linear(in_features=1280, out_features=10, bias= True))

optimiser2 = torch.optim.Adam(model2.parameters(), lr=lr, betas=(0.9, 0.999))
```

```python
# Visualize new created model
model2
```

```python
# Training loop
train(model2, optimiser2, epochs)
```

```python
# Compute test accuracy
accuracy(model2, test_loader)
```
